### Transformando Dados Categóricos em Números

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.loc[:, "default"] = LabelEncoder().fit_transform(df.default)
df.loc[:, "student"] = LabelEncoder().fit_transform(df.student)
```

```python
df.head()
```

### Separando em teste e treino

```python
X = df[["student", "balance", "income"]]
y = df.default
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Logistic Regression

### Modelando

De inicio vamos utilizar a Classe Logistic Regression com os hyperparametros default, somente ajustando os pesos das classes.

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(class_weight="balanced")
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
predicao = model.predict(X_test)
```

```python
from sklearn import metrics

pd.DataFrame(metrics.confusion_matrix(y_test, predicao))
```

```python
metrics.accuracy_score(y_test, predicao)
```

```python
print(metrics.classification_report(y_test, predicao))
```

### Avaliando o melhor modelo gerado pelo CV

```python
mmodel = LogisticRegression(C=0.5, class_weight="balanced", solver="lbfgs")
mmodel.fit(X_train, y_train)
mpredicao = mmodel.predict(X_test)
```
